```python
#@title Producto Punto de dos Matrices
def mult_matrices(A, B):
  #S = []
  s = 0
  #C = [[0,0,0], [0,0,0], [0,0,0]]

  #Inicializa Matriz resultado
  fA, cB  = len(A), len(B[0])    # Filas A, Cols B
  C = [[None] * cB for i in range(fA)]

  #COL B x FIL A
  #print("A-> F:",len(A), "C:",len(A[0]))
  #print("B-> F:",len(B), "C:",len(B[0]))
  #print("\n")

  #fA  = len(A)
  #cB  = len(B[0])
  #D = [0] * 2 # len(B[0]) # *cB
  #C = [D] * 3 #len(A)   # fA
  #C = [[0]*2] * 3 #len(A)   # fA

  for cB in range(len(B[0])): #COL B
    for fA in range(len(A)): #FIL A
      s = 0
      for cA in range(len(A[0])): #COL A
        #for fB in range(cA, cA+1): #len(B)): #FIL B
          fB = cA
          #S.append(A[fA][cA] * B[fB][cB])
          s += A[fA][cA] * B[fB][cB]
          #print(f"A[{fA}, {cA}],  {A[fA][cA]}")
          #print(f"B[{fB}, {cB}],  {B[fB][cB]}")
          #print("\n")
      #S.append("+")
      C[fA][cB] = s

  return C
```

```python
from collections.abc import Mapping

class P:
     pass

class C(P):
    pass

Mapping.register(P)
```

```python
c = C()
isinstance(c, C)        # direct
#True
isinstance(c, P)        # indirect
#True
isinstance(c, Mapping)  # virtual
#True

# Actual inheritance chain
type(c).__mro__
#(<class 'C'>, <class 'P'>, <class 'object'>)

# Test for "true inheritance"
Mapping in type(c).__mro__
#False
```

```python
import numpy as np

# Creating a 3-D numpy array using np.array()
array = np.array([[12,18,44], [66,37,89],[31,47,88]])
#array = np.array([[[12,18,44], [66,37,89]], [[31,47,88], [31,47,88]]])
print("Original array: ")
print(array)

# using = operator
D = array
D[1][1] = 2
print("\n = arr Copied array: ")
print(array, "\n")
print(D)

# using arr.copy() function
C = array.copy()
C[1][1] = 1
print("\n arr.copy() array: ")
print(array, "\n")
print(C)

# using numpy.copy() function
copy_array = np.copy(array)
print("\n np.copy() Copied array: ")
copy_array[0][0] = 0
print(array, "\n")
print(copy_array)
```
